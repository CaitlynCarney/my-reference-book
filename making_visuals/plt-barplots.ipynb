```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## How does one change the color of one bar in a matplotlib bar graph?

```python
barlist=plt.bar([1,2,3,4], [1,2,3,4])
barlist[0].set_color('r')
plt.show()
```

```python
barlist=plt.bar([1,2,3,4], [1,2,3,4])
barlist[1].set_color('r')
plt.show()
```

How about making each one different??

```python
barlist=plt.bar([1,2,3,4], [1,2,3,4])
barlist[0].set_color('skyblue')
barlist[1].set_color('lightcoral')
barlist[2].set_color('lawngreen')
barlist[3].set_color('violet')
plt.show()
```
