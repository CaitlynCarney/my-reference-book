```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## How to change the color of the axes of a plot in Matplotlib in Python

`ax.spines["bottom"].set_color("color_name")`
- this sets the x axis color
`axes.spines["left"].set_color("color_name")`
- this sets y axis color
`axes.spines["top"].set_color("color_name")`
- this sets the top of the plot
`axes.spines["right"].set_color("color_name")`
- this sets the right of the plot

## How to change the Markers on Matplotlib?

- markersize 
    - changes the size of the marker
- marker
    - changes the marker shape
- markerfacecolor
    - changes the color of the color of the marker
- markeredgecolor
    - changes the color of the markers boarders




```python
plt.plot([1, 2, 3], marker='P', markersize=40, markerfacecolor='pink', markeredgecolor='black')
```

## How to change line styles in Matplotlib

- linewidth
    - changes the width of the line
- linestyle
    - changes the style of the line
- color
    - changes color of the line



```python
plt.plot([1, 2, 3], color='green', linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=12, linewidth=5.0)
```
