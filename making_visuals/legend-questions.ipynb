## How does one change the location of their legend in 

`plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')`

- `bbox_to_anchor`
    - used to position the legend in conjunction with loc. 
        - (x, y, width, height) that the legend is placed in.

- `loc`
    - the location
        - upper left
        - best
        - etc.
