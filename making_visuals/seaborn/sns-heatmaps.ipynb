```python
from pydataset import data

import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns

import numpy as np
```

```python
df = data('iris')
```

# Making a heat map
This is fairly easy but it is where we gotta start. Then we will go on to how to adjust and customize.

```python
sns.heatmap(df.corr())
```

# How to add labels into the squares:

```python
sns.heatmap(df.corr(), annot = True)
```

### Doing this will round the correlation labels
- `fmt` allows you to set where it will round up or down too

```python
sns.heatmap(df.corr(), annot = True, fmt='.1g')
```

# How to re scale the correlation bar. 
In this example the bar is on the right hand side of the heatmap.

- `vmin` sets how low the bar will go
- `vmax` sets how high the bar will go

```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0)
```

# What if Idont want it on the side but rather below?

- `cbar_kws= {'orientation': 'horizontal'}` changes from right side to below

```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, cbar_kws= {'orientation': 'horizontal'})
```

# What if it is below but I want it on the side?

Easy:
- `cbar_kws= {'orientation': 'vertical'}`

```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, cbar_kws= {'orientation': 'vertical'})
```

# Remove the bar all together?
- you wont do this very ofter but it can come in handy sometimes
    - `cbar=False`

# Changing to color palette
Please see seaborn colors notebook for inspiration

- `cmap` is used to change the color palette in heatmaps

```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, cmap= 'mako_r')
```

# Can you add a boarder around it? If so can you change the color?
Yes and Yes
- `linewidths` allows you to create a outline and what thickness it will be
- `linecolor` sets the color it will be

```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, 
            cmap= 'mako_r', linewidths=3, linecolor='silver')
```

# If you want the square to be a perfect square rather than rectangles:
- `square=True` is your friend


```python
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, 
            cmap= 'mako_r', linewidths=3, linecolor='silver', square=True)
```

# Change the shape of the heatmap
As of right now there is not a way to outline these altered shepes without it also outlineing the removed boxes as well.

- This is helpful beacuse it basically removes all repeated and doubled up correlations

### Keep the bottom portion

```python
matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, 
            cmap= 'mako_r', square=True, mask=matrix)
```

### Keep the top portion

```python
mask = np.tril(df.corr())
sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, 
            cmap= 'mako_r', square=True, mask=mask)
```

# How can I make the bottom labels easier to read?
- Here you actually will have to set your heat map to a vlaue

```python
heatmap = sns.heatmap(df.corr(), annot = True, vmin=-1, vmax=1, center= 0, 
            cmap= 'mako_r')

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45) 
```
